```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
import pandas as pd
```

```python
ds = pd.read_csv("weatherAUS.csv")
```

```python
ds["RainToday"]=ds["RainToday"].map({"No":0,"Yes":1})
ds["RainTomorrow"]=ds["RainTomorrow"].map({"No":0,"Yes":1})
#ds = ds.apply (pd.to_numeric, errors='coerce')
ds = ds.dropna()
```

```python
ds.info()
```

```python
x=ds[["MinTemp","MaxTemp","Rainfall","WindGustSpeed","Humidity9am","Humidity3pm","Pressure9am","Pressure3pm","Temp9am","Temp3pm","RainToday"]].to_numpy()
```

```python
y=ds[["RainTomorrow"]].to_numpy()
```

```python
model = LogisticRegression(max_iter=100,random_state=0)
model.fit(x, y)
```

```python
p_pred = model.predict_proba(x)
print('p_pred:', p_pred, sep='\n', end='\n\n')
y_pred = model.predict(x)
print('y_pred:', y_pred, end='\n\n')
conf_m = confusion_matrix(y, y_pred)
print('Confusion Matrix:', conf_m, sep='\n', end='\n\n')
report = classification_report(y, y_pred)
print('Classification Report:', report, sep='\n')
```

```python
import matplotlib.pyplot as plt
a=[]
b=[]
for i in range(50,1000):
    model = LogisticRegression(max_iter=i,random_state=0)
    model.fit(x, y)
    s=model.score(x,y,sample_weight=None)
    a.append(i)
    b.append(s)
```

```python
plt.plot(a[0:100],b[0,100])
plt.xlabel('No. of iterations(max_iter)')
plt.ylabel('Accuracy')
plt.title('No. of iterations vs Accuracy')
plt.show()
```
